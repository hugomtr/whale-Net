```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import glob
import re
import os
import shutil
import pandas as pd
```

```python
try:
    os.chdir("imgs_align/")
except:
    print("already in the directory")
df = pd.read_csv("train.csv")
images_label = df.Image.values
```

```python
len(np.unique(images_label))
```

```python
images_labelises_path = df.Image.values
train_num = []
for path in images_labelises_path:
    try:
        number = re.findall("(?<=w\_)(.*)(?=\.jpg)",path).pop()
        train_num.append(number)
    except:
        print(path)
        continue
train_num
```

```python
all_jpg_names = glob.glob("wCR_*.jpg")

arr = []
for path in all_jpg_names:
    try:
        number = re.findall("(?<=wCR\_)(.*)(?=\.jpg)",path).pop()
        if number in train_num:
            shutil.copyfile(path,"FaceNetTraining/"+path)
            arr.append(number)
    except:
        continue
```

```python
set(train_num) - set(arr) 
```
