```python
df_count = df.groupby(['whaleID']).size().reset_index(name='counts')
```

```python
dist = df_count["counts"].values
fig,ax = plt.subplots(1,1,figsize = (10,10))
ax.hist(dist,bins=50)
ax.set_xlim(1,50)
ax.set_xlabel("Nombres d'apparitions")
ax.set_ylabel("Nombres d'individus")
ax.set_title("Distribution des individus en fonction du nombres d'images associées")
plt.show()
```
